```python
import pandas as pd
import numpy as np

df=pd.read_csv('NY RENTS BY YEAR.csv', index_col=0,na_values=['#DIV/0!'])
df=df.dropna(how='all')
df
```

```python


df_borough=df.groupby('Borough')['2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019', '2020'].mean().round(2)
df_borough

#df.groupby("Borough").mean()
```

```python
df_borough.to_csv('Average Rent By Borough.csv', index=True)
```
